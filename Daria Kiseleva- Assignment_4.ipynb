In order to examine wether or the relationship between two variables is really linear, it is useful to look at a *locally-smoothed regression line* that relates the $x$ and $y$ axes of a plot. This is a kind of regression model where the function is refit localy for many subsets of the data then a smooth line is interpolated between these points. One standard technique for this is known as *locally weighted scatterplot smoothing* or [LOWESS](https://en.wikipedia.org/wiki/Local_regression) and is implemented as an option for the line drawn by the `regplot` function. Let's try looking at some more complex models of the relationship between frequency and lexical decision time. Here is some starter code similar to those that we looked at in class for writing a polynomial regression. Complete the code as indicated.


```
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np

# Problem 6: Set up variable X (For features) and y from the input data.
X= young_responses['WrittenFrequency']
y=young_responses['RTlexdec']

# Problem 7: split this into test and train subsets, with 10% of the data in test.

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=1,test_size=0.1)


X_plot = np.linspace(0, 10,5000).reshape(-1, 1)
plt.scatter(X_train, y_train, color='red')

print("Model class: " + "Linear Regression")
for degree in [1,2,3,4,5,6,7,20,25]:
  # Problem 8: fit a polynomial regression model of each degree above
  r=LinearRegression()
  model=make_pipeline(PolynomialFeatures(degree), r)
  model.fit(X_train.values.reshape(-1,1), y_train)

  
  print("\tDegree " + str(degree) +"\n\t\tTrain R^2: "+ str(model.score(X_train.values.reshape(-1,1),y_train)))
  print("\t\tTest R^2: "+ str(model.score(X_test.values.reshape(-1,1),y_test)))  

  # Problem 9: use plt.plot to add a line to the plot for this model using the X_plot points.
  plt.plot(X_test.values.reshape(-1,1).ravel(), model.predict(X_test.values.reshape(-1,1)), label='degree={0}'.format(degree)) 
  plt.ylim((6, 7))
```

**Log lemma frequency** - log of the summed frequencies of all inflectional variants. More frequent verbs are less likely to be regular. 

**Log family size** log of the number of complex words in which the verb stem occurs as a constituent. Family size has no clear trend in effect on regularity.

**Auxiliary** - which auxiliary/helping verb the verb is using in present tense. If using zijnheb/have, verbs are more likely to be regular than if using zihn/be.

**Noun-verb frequency ratios** - how frequently the word appears as a noun compared to as a verb, to show verbs that are transformed from nouns. Verbs that have corresponding nouns are more likely to be regular.  

**Spoken-written frequency ratios** - how frequently the word appears in spoken corpus compared to written frequency. Verbs that are more spoken relatively often are more likely to be regular.
